```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

**from sklearn.datasets import make_classification: This line imports the make_classification function from the sklearn.datasets module. The make_classification function is a utility function in scikit-learn (sklearn) that allows you to create classification datasets with specified characteristics.
**n_samples=100: Specifies the total number of samples in the dataset. In this case, the dataset will contain 100 samples.

*n_features=2: Specifies the number of features (or independent variables) in the dataset. The dataset will have 2 features. *n_informative=1: Specifies the number of informative features, which are the features that contribute to the classification task. In this case, there is 1 informative feature.

**The resulting dataset is returned by the make_classification function and assigned to the variables X and y. X represents the feature matrix (input features) of shape (n_samples, n_features), and y represents the target vector (class labels) of shape (n_samples,).

```python
def perceptron(X,y):
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]
        
```

```python
*** what happens here is now as we have X and y as feature set so basically we have to make a line with w1,w2 weights
 and b bias or intercept now we will pick a random line and transform it into such a way that it best separates
    it linearly that is we have min no of misclassified points so this function does this job by first adding 1 to the 
    X and learning rate as 0.1 and epoch value as 1000 now we will just perform the main line to change value of weights ..
    here we will do the dot product of the input and weights and weill calculate the y_hat
    returning the intercept and the coeffice
    
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```
